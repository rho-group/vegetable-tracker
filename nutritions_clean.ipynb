```python
import pandas as pd
import numpy as np
```

```python
nutritions = pd.read_csv("raw data\\component_value.csv", sep=";")
```

```python
nutritions["EUFDNAME"] = nutritions["EUFDNAME"].fillna("NATRIUM")
```

```python
nutritions.head(40)
```

```python
nutritions.drop(columns=["ACQTYPE", "METHTYPE", "METHIND"], inplace=True)

nutritions
```

```python
df_pivot = nutritions.pivot(index="FOODID", columns="EUFDNAME", values="BESTLOC")
df_pivot
```

```python
# Delete colums which are not needed

df_pivot.drop(columns=["ENERC","FAT","CHOAVL","PROT","ALC","OA", "SUGOH","SUGAR","FRUS","GALS","GLUS","LACS","MALS","SUCS","STARCH","PSACNCS","FIBINS","FAFRE","FAPU","FAMCIS","FASAT","FATRN","FAPUN3", "FAPUN6","F18D2CN6","F18D3N3","F20D5N3","F22D6N3","CHOLE","STERT","TRP"], inplace=True)
```

```python
df_pivot
```

```python
foods = pd.read_csv("raw data\\food.csv", sep=";", encoding="iso8859_15")
foods_en = pd.read_csv("raw data\\foodname_EN.csv", sep=";", encoding="iso8859_15")
```

```python
foods = foods.merge(foods_en, on="FOODID")
```

```python
foods.drop(columns=["FOODNAME_x", "FUCLASS", "FUCLASSP", "LANG"], inplace=True)
foods
```

```python
foods = foods[["FOODID","FOODNAME_y","FOODTYPE","PROCESS","EDPORT","IGCLASS","IGCLASSP"]]

foods
```

```python
foods = foods.merge(df_pivot, on="FOODID")
```

```python
classes = ["FRUITTOT", "VEGTOT", "LEGUTOT", "POTATOT", "CERTOT"]

veg_foods = foods[foods.IGCLASSP.isin(classes)]
```

```python
veg_foods
```

```python
veg_foods.isna().sum()
```

```python
veg_foods.replace(np.nan, 0, inplace=True)
```

```python

veg_foods = veg_foods[veg_foods.PROCESS != "IND"]
```

```python
veg_foods = veg_foods[veg_foods.IGCLASS != "FBDRINK"]
veg_foods = veg_foods[veg_foods.IGCLASS != "JUICE"]
veg_foods = veg_foods[veg_foods.IGCLASS != "VEGJUICE"]
veg_foods = veg_foods[veg_foods.IGCLASS != "OATLIQ"]
veg_foods = veg_foods[veg_foods.IGCLASS != "CEROTLIQ"]
veg_foods = veg_foods[veg_foods.IGCLASS != "LEGULIQ"]
veg_foods = veg_foods[veg_foods.PROCESS != "X"]
```

```python
veg_foods
```

```python
veg_foods.columns
```

```python
list_of_columns = ['CA', 'CAROTENS', 'FE', 'FIBC', 'FOL', 'ID', 'K', 'MG',
       'NACL', 'NATRIUM', 'NIA', 'NIAEQ', 'P', 'RIBF', 'SE', 'THIA', 'VITA',
       'VITB12', 'VITC', 'VITD', 'VITE', 'VITK', 'VITPYRID', 'ZN']

for column in list_of_columns:
       veg_foods[column] = veg_foods[column].str.replace(',', '.').astype(float)
```

```python
veg_foods.to_csv("nutrition data\\nutritions.csv", index = False)
```
